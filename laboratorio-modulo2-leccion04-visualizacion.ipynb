#### Ejercicio 5: ¿Cuál es la puntuación promedio de las películas según su calificación? Deberás:

```python
df_rating_mean = pd.DataFrame(df.groupby("rating").mean(numeric_only=True)["IMDB Score"])
df_rating_mean.reset_index()
```
